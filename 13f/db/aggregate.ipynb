```python
import pandas as pd
import numpy as np
import yfinance as yf
from pandas_datareader import data as pdr
yf.pdr_override()
from datetime import timedelta
stocks_use = ['WFC', 'KO', 'AXP', 'IBM', 'WMT', 'PG', 'XOM', 'USB', 'DVA',
              'GS', 'MCO', 'GHC', 'GM', 'BK', 'COP', 'PSX', 'NOV', 'MTB', 'VRSN',
              'VZ', 'COST', 'DE', 'V', 'GL', 'LBTYA', 'MA', 'LBTYK', 'GE', 'SNY',
              'VRSK', 'MEG', 'JNJ', 'MDLZ', 'UPS', 'LEE', 'LBRDA', 'QSR', 'FWONK',
              'DNOW', 'KHC', 'AAPL', 'AXTA', 'KMI', 'LILA', 'LILAK', 'CHTR', 'LUV',
              'DAL', 'AAL', 'UAL', 'SIRI', 'BAC', 'SYF', 'TEVA', 'STOR', 'JPM', 'PNC',
              'AMZN', 'TRV', 'STNE', 'KR', 'OXY', 'BIIB', 'RH', 'CVX', 'ABBV', 'BMY',
              'SNOW', 'MRK', 'AON', 'TMUS', 'MMC', 'ATVI', 'HPQ', 'C', 'PARA', 'CE',
              'MCK', 'MKL', 'ALLY', 'FND', 'RPRX', 'FWONA', 'OGN', 'T', 'ORCL', 'PFE', 'TSM', 'LPX', 'JEF']
```

```python
df_merged = pd.DataFrame(columns=['Sym', 'Amount'])
```

```python
institution = 'Berkshire Hathaway Inc'
quarters = ['Q1','Q2','Q3','Q4']
years = [str(y) for y in range(2013,2023)]
for y in years:
    for q in quarters:
        try:
            df = pd.read_csv(f'./people/{institution}/{institution} {q} {y} 13F Top Portfolio Holdings.csv')
            for i in range(len(df)):
                df['Value ($000)'][i]=df['Value ($000)'].values[i].replace(',','')
            df['Value ($000)'] = df['Value ($000)'].astype(float)
            df = df.drop(['Cl','CUSIP','%','Principal','Option Type','Issuer Name','Shares'], axis=1)
        
            quarter_dict = {'Q1':'03-31','Q2':'06-30','Q3':'09-30','Q4':'12-31'}
            day = y+'-'+quarter_dict[q]
            dday = pd.to_datetime(day)
            # NumShares = []
            for i in range(len(df)):
                try:
                    ticker = df['Sym'][i]
                    price = pdr.get_data_yahoo(ticker,start=day, end = dday + timedelta(3))['Close'][0]
                    df['Value ($000)'][i] = round(df['Value ($000)'][i]/price,1)
                    #NumShares.append(np.round(amt,1))
                    
                except:
                    print('shat')
                    pass
            df.rename(columns={'Value ($000)':'Amount (000)'}, inplace=True)
            
            for i in range(len(df)):
                if df['Sym'][i] not in stocks_use:
                    df.drop(i, axis=0,inplace=True)
            df.reset_index(inplace=True)
            del df['index']
            
            cnt = len(df)
            for s in stocks_use:
                if s not in list(df['Sym']):
                    df.loc[cnt] = [s, 0]
                    cnt += 1
            if len(df) != 90:
                print(f'Error for {y} {q}')
                break;
            
            quarter_dict2 = {'Q1':'03-01','Q2':'06-01','Q3':'09-01','Q4':'12-01'}
            df['date'] = pd.to_datetime(f'{y}-{quarter_dict2[q]}')
            df = df.set_index('date')
            
            df_merged = pd.concat([df_merged,df],axis=0)
            print(f'Merged {y} {q}!')
        except:
            print(f'Error with Loading Data {y} {q}..')
            pass
            
```
